Step 4: Data cleaning and viz

```python
# SET INDEX
df.set_index("DATE", inplace= True)
# df = df.set_index("DATE")
df
```

```python
# Visualization
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.title("Real GDP")
plt.xlabel("Date")
plt.ylabel("Billions of dollar")
plt.grid()
plt.plot(df.GDP)
plt.show()
```
